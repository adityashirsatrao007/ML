```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load an external Kaggle dataset: Netflix Movies and TV Shows
df = pd.read_csv('netflix_titles.csv')

# Display DataFrame structure
print(df.head())

# Pandas: Counting movies vs. TV shows
content_counts = df['type'].value_counts()
print("\nContent Type Counts:\n", content_counts)

# NumPy: Calculate average release year (for demonstration)
release_years = df['release_year'].dropna().astype(np.int32)
print("Average Release Year (NumPy):", np.mean(release_years))

# Example Data Cleaning: Fill missing country values
df['country'] = df['country'].fillna('Unknown')

# Scikit-learn: Encode country column
encoder = LabelEncoder()
df['country_encoded'] = encoder.fit_transform(df['country'])

# Bar Chart: Top 5 genres by count
plt.figure(figsize=(10, 6))
genre_counts = df['listed_in'].value_counts().head(5)
plt.bar(genre_counts.index, genre_counts.values, color='skyblue')
plt.title('Top 5 Content Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# Pie Chart: Movies vs. TV Shows
plt.figure(figsize=(8, 8))
plt.pie(content_counts, labels=content_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution: Movies vs TV Shows')
plt.show()

# Box Plot: Release years for Movies
plt.figure(figsize=(8, 6))
df[df['type'] == 'Movie'].boxplot(column='release_year', vert=False)
plt.title('Release Year Distribution for Movies')
plt.xlabel('Release Year')
plt.show()

# Histogram: Frequency of Titles by Release Year (last 20 years)
plt.figure(figsize=(10, 6))
last_20_years = df[df['release_year'] >= (df['release_year'].max() - 20)]
plt.hist(last_20_years['release_year'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Title Releases per Year (Last 20 Years)')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()

# Line Chart & Subplots: Number of titles over years (Movies vs TV Shows)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
movies_per_year = df[df['type'] == 'Movie']['release_year'].value_counts().sort_index()
tv_per_year = df[df['type'] == 'TV Show']['release_year'].value_counts().sort_index()

axes[0].plot(movies_per_year.index, movies_per_year.values, color='green')
axes[0].set_title('Movies Released per Year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Count')

axes[1].plot(tv_per_year.index, tv_per_year.values, color='orange')
axes[1].set_title('TV Shows Released per Year')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
