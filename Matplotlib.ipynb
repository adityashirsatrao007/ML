```python
#
# Matplotlib Workflow: All-in-One Code Block (Final Correction)
# This script is corrected to handle the modern behavior of get_sample_data.
#

#
# 1. IMPORTS
#
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cbook import get_sample_data

#
# 2. PREPARE THE DATA
#
print("--- 1. Preparing Data ---")
# 1D Data
x_1d = np.linspace(0, 10, 100)
y_1d = np.cos(x_1d)
z_1d = np.sin(x_1d)

# 2D Data or Images
data_2d = 2 * np.random.random((10, 10))
data_2d_2 = 3 * np.random.random((10, 10))
# CORRECTED: get_sample_data now directly returns the NumPy array.
img = get_sample_data('axes_grid/bivariate_normal.npy')

# Vector Field Data
Y_vec, X_vec = np.mgrid[-3:3:100j, -3:3:100j]
U_vec = -1 - X_vec**2 + Y_vec
V_vec = 1 + X_vec - Y_vec**2
print("Data prepared successfully.\n")


#
# 3. CREATE PLOT (FIGURES AND AXES)
#
print("--- 2. Creating Plot Objects ---")
# Create a new figure
fig = plt.figure()
# Create a figure with a specific size and aspect ratio
fig2 = plt.figure(figsize=plt.figaspect(2.0))

# Add an axes to the figure
ax = fig.add_subplot(111)

# Create a figure and a grid of subplots
fig3, axes = plt.subplots(nrows=2, ncols=2)
# Create a figure and a grid with 3 columns
fig4, axes2 = plt.subplots(ncols=3)
print("Figure and Axes objects created.\n")

#
# 4. PLOTTING ROUTINES
#
print("--- 3. Populating Axes with Plots ---")

# --- Plotting 1D Data ---
fig_1d, axes_1d = plt.subplots(2, 3, figsize=(12, 7))
plt.suptitle("Plotting Routines for 1D Data")
lines = axes_1d[0, 0].plot(x_1d, y_1d)
axes_1d[0, 0].set_title("plot")
axes_1d[0, 1].scatter(x_1d, y_1d)
axes_1d[0, 1].set_title("scatter")
axes_1d[0, 2].bar([1, 2, 3], [3, 4, 5])
axes_1d[0, 2].set_title("bar")
axes_1d[1, 0].barh([0.5, 1, 2.5], [0, 1, 2])
axes_1d[1, 0].set_title("barh")
axes_1d[1, 1].axhline(0.45)
axes_1d[1, 1].axvline(0.65)
axes_1d[1, 1].set_title("axhline & axvline")
axes_1d[1, 2].fill(x_1d, y_1d, color='blue')
axes_1d[1, 2].fill_between(x_1d, y_1d, color='yellow')
axes_1d[1, 2].set_title("fill & fill_between")
fig_1d.tight_layout(rect=[0, 0.03, 1, 0.95])

# --- Plotting 2D Data ---
fig_2d, axes_2d = plt.subplots(1, 3, figsize=(12, 4))
plt.suptitle("Plotting Routines for 2D Data")
im = axes_2d[0].imshow(img, cmap='gist_earth')
axes_2d[0].set_title("imshow")
axes_2d[1].pcolor(data_2d)
axes_2d[1].set_title("pcolor")
CS = axes_2d[2].contour(Y_vec, X_vec, U_vec)
axes_2d[2].clabel(CS)
axes_2d[2].set_title("contour")
fig_2d.tight_layout(rect=[0, 0.03, 1, 0.95])


# --- Plotting Data Distributions ---
fig_dist, axes_dist = plt.subplots(1, 3, figsize=(12, 4))
plt.suptitle("Plotting Data Distributions")
axes_dist[0].hist(y_1d)
axes_dist[0].set_title("hist")
axes_dist[1].boxplot(y_1d)
axes_dist[1].set_title("boxplot")
axes_dist[2].violinplot(z_1d)
axes_dist[2].set_title("violinplot")
fig_dist.tight_layout(rect=[0, 0.03, 1, 0.95])

# --- Plotting Vector Fields ---
fig_vec, axes_vec = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle("Plotting Vector Fields")
axes_vec[0].quiver(x_1d[::5], y_1d[::5])
axes_vec[0].set_title("quiver")
axes_vec[1].streamplot(X_vec, Y_vec, U_vec, V_vec)
axes_vec[1].set_title("streamplot")
fig_vec.tight_layout(rect=[0, 0.03, 1, 0.95])


#
# 5. CUSTOMIZE PLOT
#
print("--- 4. Customizing Plot ---")
fig_cust, ax_cust = plt.subplots(figsize=(8, 6))

line, = ax_cust.plot(x_1d, y_1d, color='darkgreen', marker='o', linestyle='--', linewidth=2.0, alpha=0.8)
plt.setp(line, color='r', linewidth=4.0)

ax_cust.set(title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')
plt.title(r'$\sigma_i=15$', fontsize=20)

ax_cust.text(1, -0.5, 'Example Text', style='italic')
ax_cust.annotate("Peak", xy=(np.pi, 1), xytext=(4, 1.25), arrowprops=dict(arrowstyle="->"))

ax_cust.set_xlim(0, 10.5)
ax_cust.set_ylim(-2.0, 2.0)

ax_cust.xaxis.set(ticks=range(1, 10), ticklabels=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'])
ax_cust.tick_params(axis='y', direction='inout', length=10)

ax_cust.plot(x_1d, z_1d, label='Sine')
ax_cust.legend(loc='best')

ax_cust.spines['top'].set_visible(False)
ax_cust.spines['bottom'].set_position(('outward', 10))

fig_cust.tight_layout()
print("Customizations applied.\n")

#
# 6. SAVE AND SHOW PLOT
#
print("--- 5 & 6. Saving and Showing Plot ---")

# Save the plot to a file
# plt.savefig('my_custom_plot.png')
print("Plot can be saved using 'plt.savefig()'.")

# Show all the created plot windows
plt.show()

print("Script finished.")
```
