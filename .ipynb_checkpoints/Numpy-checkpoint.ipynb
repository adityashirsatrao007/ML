#INDEXING/SLICING/SUBSETTING

```python
new_arr2d = np.array([[10, 20, 30],
                      [40, 50, 60],
                      [70, 80, 90]])
```

```python
new_arr2d[2, 1]  # Returns the element at row 2, column 1 (0-based index)
```

```python
new_arr2d[1, 2] = 100  # Assigns 100 to element at row 1, column 2
```

```python
new_arr2d[0:2, 1]  # Returns elements from rows 0 to 1 of column 1
```

```python
new_arr2d[:, 1]  # Returns all elements of column 1
```

```python
new_arr = np.array([10, 20, 30, 40, 50, 60, 70])
```

```python
new_arr[5]
```

```python
new_arr[1] = 4  # Assigns array element at index 1 the value 4
```

```python
new_arr[0:3]  # Returns the elements at indices 0,1,2
```

```python
new_arr[:2]  # Returns the elements at indices 0,1
```

```python
new_arr < 50  # Returns a boolean array for elements less than 50
```

```python
new_arr[new_arr < 50]  # Returns array elements smaller than 50
```

#VECTOR MATH

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([5, 4, 3, 2, 1])
print(arr1,arr2)
```

```python
np.add(arr1, arr2)
```

```python
np.subtract(arr1, arr2)
```

```python
np.multiply(arr1, arr2) 
```

```python
np.divide(arr1, arr2)
```

```python
np.power(arr1, arr2)
```

```python
np.array_equal(arr1, arr2)
```

```python
np.sqrt(arr1)
```

```python
np.sin(arr1)
```

```python
np.log(arr1)
```

```python
np.ceil(arr1 / 2.5)
```

```python
np.floor(arr1 / 2.5)
```

```python
np.round(arr1 / 2.5) 
```
