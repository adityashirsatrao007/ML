#EXPORTING DATA

```python
data = {
    "Name": ["Aditya", "Srushti", "Viha"],
    "Age": [22, 21, 23],
    "City": ["Solapur", "Pune", "Mumbai"]
}
df = pd.DataFrame(data)

# Write to CSV
df.to_csv("file.csv", index=False)
```

```python
# Write to Excel
df.to_excel("SIH_PS_2024.xlsx", index=False)
```

```python
import sqlite3

# Connect to database
conn = sqlite3.connect("sample.db")

# Write DataFrame to SQL table
df.to_sql("Sample_SQL_Table", conn, if_exists="replace", index=False)

# Close connection
conn.close()
```

```python
# Write to JSON
df.to_json("iris.json", orient="records", lines=True)
```

```python
# Write to HTML
df.to_html("table-download.html", index=False)
```

```python
# Copy DataFrame to clipboard
df.to_clipboard(index=False)
```

#JOIN/COMBINE

```python
df1 = pd.DataFrame(np.random.rand(3, 3), columns=['a', 'b', 'c'])
df2 = pd.DataFrame(np.random.rand(2, 3), columns=['a', 'b', 'c'])
```

```python
df_combined = pd.concat([df1, df2], axis=0, ignore_index=True)
df_combined
```

```python
pd.concat([df1, df2], axis=1)
```

```python
df_joined = df1.join(df2.set_index('a'), on='a', how='inner', lsuffix='_df1', rsuffix='_df2')
df_joined
```
