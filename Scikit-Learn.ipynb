```python
#
# Scikit-Learn Workflow: All-in-One Code Block (Corrected for ValueError)
# This script demonstrates a complete, step-by-step machine learning workflow
# using the scikit-learn library, from data loading to model tuning.
#

#
# 1. IMPORTS
#
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, Normalizer, Binarizer, LabelEncoder, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn import neighbors
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix,
                             mean_absolute_error, mean_squared_error, r2_score,
                             adjusted_rand_score, homogeneity_score, v_measure_score)
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

#
# 2. LOADING THE DATA
#
X = np.random.rand(10, 5)
y_categorical = np.array(['M', 'M', 'F', 'F', 'M', 'F', 'M', 'M', 'F', 'F'])

#
# 3. PREPROCESSING THE DATA
#
print("--- PREPROCESSING ---")
enc = LabelEncoder()
y = enc.fit_transform(y_categorical)
print(f"Original categorical labels: {y_categorical}")
print(f"Encoded integer labels: {y}\n")

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(f"Data split into {X_train.shape[0]} training samples and {X_test.shape[0]} test samples.\n")

scaler = StandardScaler().fit(X_train)
standardized_X_train = scaler.transform(X_train)
standardized_X_test = scaler.transform(X_test)
print("Standardization applied.\n")

normalizer = Normalizer().fit(X_train)
normalized_X_train = normalizer.transform(X_train)
normalized_X_test = normalizer.transform(X_test)
print("Normalization applied.\n")

binarizer = Binarizer(threshold=0.5).fit(X)
binary_X = binarizer.transform(X)
print("Binarization applied.\n")

X_train_with_missing = X_train.copy()
X_train_with_missing[0, 0] = 0
imp = SimpleImputer(missing_values=0, strategy='mean')
imputed_X_train = imp.fit_transform(X_train_with_missing)
print("Imputation of missing values complete.\n")

poly = PolynomialFeatures(degree=2)
polynomial_X = poly.fit_transform(X)
print(f"Generated polynomial features. Original shape: {X.shape}, New shape: {polynomial_X.shape}\n")


#
# 4. CREATING AND FITTING MODELS
#
print("--- MODEL TRAINING ---")
knn = neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("KNN model fitted.")

svc = SVC(kernel='linear', probability=True)
svc.fit(X_train, y_train)
print("SVC model fitted.")

k_means = KMeans(n_clusters=2, random_state=0)
k_means.fit(X_train)
print("K-Means model fitted.")

pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train)
print(f"PCA model fitted and data transformed. New shape: {X_train_pca.shape}\n")


#
# 5. MAKING PREDICTIONS
#
print("--- PREDICTIONS ---")
y_pred_knn = knn.predict(X_test)
y_pred_svc = svc.predict(X_test)
y_pred_kmeans = k_means.predict(X_test)
print(f"KNN Predictions: {y_pred_knn}")
print(f"SVC Predictions: {y_pred_svc}")
print(f"K-Means Predictions: {y_pred_kmeans}\n")

y_proba_knn = knn.predict_proba(X_test)
print(f"KNN Class Probabilities (first sample): {y_proba_knn[0]}\n")

#
# 6. EVALUATING MODEL PERFORMANCE
#
print("--- MODEL EVALUATION ---")
print("--- Classification ---")
print(f"Accuracy Score (KNN): {accuracy_score(y_test, y_pred_knn):.2f}")
print(f"Estimator Score (SVC): {svc.score(X_test, y_test):.2f}\n")
print("Classification Report (SVC):")
print(classification_report(y_test, y_pred_svc, zero_division=0))
print("Confusion Matrix (KNN):")
print(confusion_matrix(y_test, y_pred_knn))

print("\n--- Regression (Example) ---")
y_true_reg = np.array([3, -0.5, 2, 7])
y_pred_reg = np.array([2.5, 0.0, 2, 8])
print(f"Mean Absolute Error: {mean_absolute_error(y_true_reg, y_pred_reg)}")
print(f"Mean Squared Error: {mean_squared_error(y_true_reg, y_pred_reg)}")
print(f"R2 Score: {r2_score(y_true_reg, y_pred_reg)}\n")

print("--- Clustering ---")
print(f"Adjusted Rand Score: {adjusted_rand_score(y_test, y_pred_kmeans):.2f}")
print(f"Homogeneity Score: {homogeneity_score(y_test, y_pred_kmeans):.2f}\n")

print("--- Cross-Validation ---")
lr = LinearRegression()
cv_scores = cross_val_score(lr, X, y, cv=3)
print(f"Cross-Validation Scores (Linear Regression): {cv_scores}")
print(f"Average CV Score: {cv_scores.mean():.2f}\n")

#
# 7. TUNING THE MODEL
#
print("--- MODEL TUNING ---")
# CORRECTED: Added cv=3 to handle the small dataset size
params_grid = {"n_neighbors": np.arange(1, 3), "metric": ["euclidean", "cityblock"]}
grid = GridSearchCV(estimator=knn, param_grid=params_grid, cv=3)
grid.fit(X_train, y_train)
print(f"Grid Search Best Score: {grid.best_score_:.2f}")
print(f"Grid Search Best Parameters: {grid.best_params_}\n")

# CORRECTED: Added cv=3 to handle the small dataset size
params_rand = {"n_neighbors": range(1, 5), "weights": ["uniform", "distance"]}
rsearch = RandomizedSearchCV(estimator=knn, param_distributions=params_rand, n_iter=4, random_state=5, cv=3)
rsearch.fit(X_train, y_train)
print(f"Randomized Search Best Score: {rsearch.best_score_:.2f}")
print(f"Randomized Search Best Parameters: {rsearch.best_params_}")
```
